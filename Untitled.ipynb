## Degree centrality

- Degree сentrality metric defines importance of a node in a graph as being measured based on its degree.
- It tells us **how many direct connections** each node has to other nodes in the network.
- Used to find very connected individuals, popular individuals, good candidates **for message translation and distribution of information**.

```python
degree_centrality_df = pd.DataFrame(nx.degree_centrality(network).items(), 
                                       columns=['node', 'degree_centrality'])
degree_top_5 = degree_centrality_df.sort_values('degree_centrality', ascending=False).head(5)
degree_top_5
```

```python
degree_centrality_df.degree_centrality.hist();
```

```python
highlight_top_central_nodes(network, degree_top_5.node.values)
```

Speaking in terms of social network, people as `nodes 107, 1684, 1912 and 3437` know most of people (they have the most direct connections with other nodes).

## Eigenvector centrality

- Eigenvector centrality can identify nodes with influence over the whole network by calculating **the extended connections of a node (looking at the neighbors)**.
- Used to find individuals **for change management**.

```python
eigenvector_centrality_df = pd.DataFrame(nx.eigenvector_centrality(network).items(), 
                                       columns=['node', 'eigenvector_centrality'])
eigenvector_top_5 = eigenvector_centrality_df.sort_values('eigenvector_centrality', ascending=False).head(5)
eigenvector_top_5
```

```python
eigenvector_centrality_df.eigenvector_centrality.hist();
```

```python
highlight_top_central_nodes(network, eigenvector_top_5.node.values)
```

Eigenvector centrality is about **it’s not how many you know, but who you know**. In a social network **you may have more influence if you know a few influential people than if you know many not-so-influential people**. <br>
We see that `node 1912` is central according to both degree and eigenvector centrality calculations, however there are many differences in ranking other nodes. It may be explained looking at the picture below:

## PageRank centrality

- PageRank is a variant of Eigenvector centrality. PageRank’s main difference from EigenCentrality is that it **accounts for link direction**. Each node in a network is assigned a score based on its number of incoming links.
- As it takes into account direction and connection weight, it can be used for **finding influential people and understand their authority**.

```python
pagerank_centrality_df = pd.DataFrame(nx.pagerank(network).items(), 
                                       columns=['node', 'pagerank_centrality'])
pagerank_top_5 = pagerank_centrality_df.sort_values('pagerank_centrality', ascending=False).head(5)
pagerank_top_5
```

```python
pagerank_centrality_df.pagerank_centrality.hist();
```

```python
highlight_top_central_nodes(network, pagerank_top_5.node.values)
```

We observe the same nodes as after degree centrality calculation being applied. It may be explained by the fact that the PageRank algorithm was designed for directed graphs, but in our case for undirectedd graph it converted each edge in the directed graph to two edges and the calculations are simplified to the ones that are in degree centrality.

## Unconstraint centrality

- Based on constraint calculation, which is a measure of **structural holes**. 
- **A structural hole** is understood as a gap between two individuals who have complementary sources to information. 
- It tells us **how much new connections you open for your neighbors**.
- Used for building of **bridges** between locations/units and for cross-functional projects. An individual who acts as a mediator between two or more closely connected groups of people could gain important **comparative advantages** llike transfering or keeping valuable information, combining ideas.
- The **constraint** is a measure of the extent to which a node *v* is invested in those nodes that are themselves invested in the neighbors of *v*.

More info about structural holes and its measures you may find [here](https://en.wikipedia.org/wiki/Structural_holes).

```python
unconstraint_centrality_df = pd.DataFrame(nx.algorithms.structuralholes.constraint(network).items(),
                                          columns=['node', 'unconstraint_centrality'])
```

```python
unconstraint_top_5 = unconstraint_centrality_df.sort_values('unconstraint_centrality').head(5)
unconstraint_top_5
```

```python
unconstraint_centrality_df.unconstraint_centrality.hist();
```

```python
highlight_top_central_nodes(network, unconstraint_top_5.node.values)
```

Again we observe the same nodes as after degree centrality calculation being applied. 
This time we are interested in the smallest values because high constraint means inefficient nodes (they invest time and energy in the relation to those whose network lacks structural holes and is also tied to other contacts in the person’s network). A low constraint value indicates a person who doesn't have many direct connections with other people but may serve as a bridge between two or more clusters.

## Closeness centrality

- It calculates the shortest paths between all nodes, then assigns each node a score based on its sum of shortest paths.
- Used to find the individuals **who are best placed to influence the entire network most quickly**.

```python
closeness_centrality_df = pd.DataFrame(nx.closeness_centrality(network).items(), 
                                       columns=['node', 'closeness_centrality'])
closeness_top_5 = closeness_centrality_df.sort_values('closeness_centrality', ascending=False).head(5)
closeness_top_5
```

```python
closeness_centrality_df.closeness_centrality.hist();
```

```python
highlight_top_central_nodes(network, closeness_top_5.node.values)
```

`Nodes 107, 58, 428, 563, 1684` are closest to all other nodes in the network in terms of geodesic paths.

## Betweenness centrality

who influence the flow around a system.- It measures the number of times a node lies on the shortest path between other nodes.
- This measure **shows which nodes are ‘bridges’ between nodes in a network**. It does this by identifying all the shortest paths and then counting how many times each node falls on one.
- Useto find the individuals **who influence the flow around a system**.

```python
betweenness_centrality_df = pd.DataFrame(nx.betweenness_centrality(network).items(), 
                                       columns=['node', 'betweenness_centrality'])
betweeness_top_5 = betweenness_centrality_df.sort_values('betweenness_centrality', ascending=False).head(5)
betweeness_top_5
```

```python
betweenness_centrality_df.betweenness_centrality.hist();
```

```python
highlight_top_central_nodes(network, betweeness_top_5.node.values)
```

Betweenness centrality also shows us which nodes may serve as bridges so we see most of the same nodes as unconstraint centraity calculations are being applied.

## Centrality Summary

```python
centrality_summary = degree_centrality_df\
    .merge(eigenvector_centrality_df, on='node')\
    .merge(closeness_centrality_df, on='node')\
    .merge(betweenness_centrality_df, on='node')\
    .merge(pagerank_centrality_df, on='node')
    # .merge(unconstraint_centrality_df, on='node')\


centrality_summary
```

### Ranking centrality

```python
centrality_summary.set_index('node', inplace=True)
centrality_ranks = centrality_summary.rank(ascending=False, 
                                           numeric_only=True, 
                                           method='dense').astype(int)
centrality_ranks.sort_values('degree_centrality')
```

```python
uncunconstraint_centrality_df_index = unconstraint_centrality_df.set_index('node')
unconstraint_ranks = uncunconstraint_centrality_df_index.rank(ascending=True, 
                                                              numeric_only=True, 
                                                              method='dense').astype(int)

unconstraint_ranks.sort_values('unconstraint_centrality')
```

```python
centrality_all_ranks = centrality_ranks.join(unconstraint_ranks)
```

### Average ranking

```python
centrality_all_ranks['average_rank'] = round(centrality_all_ranks.mean(axis=1)).astype(int)
centrality_all_ranks.sort_values('average_rank')
```

```python
centrality_all_ranks.average_rank.hist();
```
